```python
from SC import scsimulation
import matplotlib.pyplot as plt
```

```python
# Parameter Settings
wavelength = 1960  # Wavelength (nm)
t0 = 0.06          # Pulse Width (ps)
power = 800        # Peak Power (W)
flength = 23.77    # Fiber Length (mm)
loss = 400         # Loss (dB/m)
C = 0              # Initial Chirp
betas = [-3.9629261e-01, 3.9372774e-03, -9.7895932e-06, 7.1608533e-08, -1.1079456e-10, -6.1391192e-12,
         1.3216948e-14, 5.1218183e-16] # ps^n/m
gamma = 20.7248      # 1/W/m
```

```python
model = scsimulation.Simulation(wavelength=wavelength, t0=t0, power=power, flength=flength, loss=loss, C=C, betas=betas, gamma=gamma)
model.sech()  # 使用双曲正割
# model.gauss() # 使用高斯型
model.gnlse_cal(rtol=1e-5, atol=1e-6) # 设置求解精度
plt.rcParams["figure.figsize"] = (16, 10) # 设置图像大小
model.gnlse_plot() # 如果输出了两个txt文件，表示计算结束。绘图时间比较长，得耐心等待
```

```python
model = scsimulation.Simulation(wavelength=wavelength, t0=t0, power=power, flength=flength, loss=loss, C=C, betas=betas, gamma=gamma)
# model.sech()  # 使用双曲正割
model.gauss() # 使用高斯型
model.gnlse_cal(rtol=1e-5, atol=1e-6) # 设置求解精度
plt.rcParams["figure.figsize"] = (16, 10) # 设置图像大小
model.gnlse_plot() # 如果输出了两个txt文件，表示计算结束。绘图时间比较长，得耐心等待
```
